```python
import pandas as pd, ast, gzip # Se importan las librerías necesarias.
```

En el siguiente código, se extrae del archivo "user_items.json.gz" y se convierte a un dataframe de pandas.

```python
data = []

with gzip.open(r"C:\Users\roylo\Downloads\PI MLOps - STEAM-20240520T190249Z-001\PI MLOps - STEAM\users_items.json.gz", 'rt',encoding='utf-8') as file:
    for line in file:
        try:
            json_data = ast.literal_eval(line)
            data.append(json_data)
        except ValueError as e:
            print(f"Error en la línea: {line}")
            continue

user_items = pd.DataFrame(data) 

user_items.head()
```

Observando la variable "items" se puede notar que es una columna anidada, por lo que en el siguiente código se construye una función que ayude a limpiarla.

```python
def CleanItems(l:list)->dict:
    """
    Ésta función extrae los valores de la lista de diccionarios, la cual es la variable original. Esto lo hace iterando, extrayendo el respectivo valor 
    y añadiendolo a un nuevo diccionario con los valores en una lista. En pocas palabras, se convierte una lista de diccionarios a un solo diccionario 
    con los respectivos valores en dos listas.
    """
    dic = {'item_id':[],
           'item_name':[]}
    
    for i in l:
        dic['item_id'].append(i.get('item_id'))
        dic['item_name'].append(i.get('item_name'))
    
    return dic

user_items['items2'] = user_items['items'].apply(CleanItems)  

items = user_items.pop('items') # Se extrae la variable original y se guarda en una variabe.

user_items.head()
```

De manera de seguir limpiando la variable "items2", se itera la variable extrayendo sus dos valores y insertandolos en dos respectivas listas, creando una lista de listas. Después, se definen nuevas variables con los datos de dichas listas.

```python
items_id = []

items_name = []

for i in user_items['items2']:
    items_id.append(i.get('item_id'))
    items_name.append(i.get('item_name'))

items2 = user_items.pop('items2') # Se extrae y se guarda en una variable.

user_items['item_id'] = items_id

user_items['item_name'] = items_name

user_items.head()
```
