Lo que sigue, es explorar las demas variables; limpiarlas en caso de ser necesario y normalizarlas.

```python
steam_games['release_date'] = pd.to_datetime(steam_games['release_date'],format='mixed',errors='coerce') # Con "pd.to_datetime" se transforma a la variable "release_date". En el parámetro "errors", "coerce" permite 
                                                                                                         # manejar los errores como "Nan".

steam_games.dropna(inplace=True)
steam_games.reset_index(drop=True,inplace=True)

steam_games['release_date'].head()
```

```python
steam_games['price'].unique() # Se exploran los valores únicos de la variable "price"
```

Con la previa exploración se pudo notar que hay varios valores que hacen referencia a un producto gratuito y otros que no. Para determinar si un valor es gratuito o no, en éste contexto se buscaron las palabras clave "Free y Demo"; cualquier otro valor no numerico se considera nulo.

```python
free = ['Free To Play','Free to Play','Free','Free Demo','Play for Free!','Play WARMACHINE: Tactics Demo','Free Mod','Free HITMAN™ Holiday Pack','Play the Demo'] # Valores que se identificaron como gratuitos.

def SetFree(s :str)->str:
    """
    Con ésta función se busca limpiar la variable "price".
        input: Es un valor de tipo "str".

        output: Es "0" si se encuentra en la lista "free".  De lo contrario devuelve el valor original.
    """
    if s in free:
        return '0'
    else:
        return s
    
steam_games['price'] = steam_games['price'].apply(SetFree)  # Se aplica la función "SetFree".

steam_games['price'] = pd.to_numeric(steam_games['price'],errors='coerce') # Con "pd.to_numeric" se transforma la variable "price" un valor flotante, donde "errors='coerce'" permite el manejo de errores transformandolos a NaN.

steam_games.dropna(inplace=True) # Se eliminan los valores nulos y se resetea el índice.
steam_games.reset_index(drop=True, inplace=True)

steam_games.head()
```

```python
steam_games['id'] = steam_games['id'].astype(int) # Se transforma la variable "id" a int.
```

```python
def CleanStr(s:str)->str:       # En caso de que haya valores con espacios de más.
    """
    Ésta función busca limpiar los espacios a los lados.
        input: Es un valor de tipo "str".

        output: Es un valor de tipo "str".
    """
    s = re.sub(r"\s+",' ',s).strip()
    return s

steam_games['publisher'] = steam_games['publisher'].apply(CleanStr)

steam_games['app_name'] = steam_games['app_name'].apply(CleanStr)

steam_games['title'] = steam_games['title'].apply(CleanStr)

steam_games['developer'] = steam_games['developer'].apply(CleanStr)

steam_games.head()
```
