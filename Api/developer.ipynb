```python
import pandas as pd
```

```python
# Se importan sólo las columnas necesarias buscando optimizar la consulta.

developer = pd.read_parquet(r'C:\Users\roylo\OneDrive\Documentos\Data Science\Proyectos Individuales\Machine Learning Operations\Datasets\steam_games.parquet',columns=['release_date','price','developer'])

developer.head()
```

```python
# Se extrae el año de la variable "release_date"

developer['release_date'] = developer['release_date'].dt.year

developer.head()
```

```python
# Se cuentas los items por año y por desarrollador.

itemscount = developer.groupby(by=['release_date','developer']).count().rename(columns={'price':'ItemsCount','release_date':'Año'})

itemscount.head()
```

```python
# Se cuentan los productos gratuitos por año y por desarrollador.

freecount = developer[developer['price']==0].groupby(by=['release_date','developer']).count().rename(columns={'price':'freecount','release_date':'Año'})

freecount.head()
```

```python
# Se unen los dos datasets previamente creados, se llenan los valores nulos con 0 en casos donde no hay contenido gratuito.

developer = pd.concat([itemscount,freecount['freecount']],axis=1).fillna(0).reset_index().rename(columns={'release_date':'Año','ItemsCount':'Cantidad de Items'})

developer['Contenido Free'] = developer['freecount']/developer['Cantidad de Items']*100 # Se obtiene el porcentaje del contenido gratuito respecto al total de items.

developer.drop(columns='freecount',inplace=True)

developer.set_index('Año',inplace=True)

developer.head()
```

```python
developer['Contenido Free'] = developer['Contenido Free'].astype(int).astype(str) # Se redondea la variable price para después convertirla en "str".

developer['Contenido Free'] = developer['Contenido Free'] + "%"

developer.head()
```
